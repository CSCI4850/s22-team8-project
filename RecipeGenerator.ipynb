```python
import random
import numpy

# Declare lists of constants for ingredient generation

ing_averages = []
ing_stdev = []

ing_averages.append(1)
ing_averages.append(59.5)
ing_averages.append(2.1)
ing_averages.append(73.3)
ing_averages.append(3.8)
ing_averages.append(2.3)
ing_averages.append(9.8)
ing_averages.append(1.3)
ing_averages.append(75)
ing_averages.append(6.4)
ing_averages.append(3.5)
ing_averages.append(2.5)
ing_averages.append(7.9)
ing_averages.append(100.6)
ing_averages.append(37.5)

ing_stdev.append(0)
ing_stdev.append(35.9)
ing_stdev.append(1.0)
ing_stdev.append(21.8)
ing_stdev.append(1)
ing_stdev.append(1.3)
ing_stdev.append(1.4)
ing_stdev.append(0.3)
ing_stdev.append(0.5)
ing_stdev.append(3.4)
ing_stdev.append(0.5)
ing_stdev.append(0.5)
ing_stdev.append(0.5)
ing_stdev.append(56.7)
ing_stdev.append(5.4)
```

```python
def generateRecipe(verbose = False):
    ingredients = []

    # Initialize all ingredients to zero (i.e. not used in recipe) except yeast
    ingredients.append(1)
    for i in range(0,14):
        ingredients.append(0)

    # Which liquids?
    r = random.random()
    if r < .2:
        ingredients[14] = 1 # 1/5 chance of milk
        
    if r >= .1:
        ingredients[1] = 1 # 9/10 chance of water
                            # small chance of both OK (.1 >= r < .2)
    if (verbose):
        print("Liquid determination: ", r)
            
    # Which flour?
    r = random.random()    
    if r < .6:
        ingredients[3] = 1
    elif r < .9:
        ingredients[13] = 1
    else:
        ingredients[8] = 1
        
    if (verbose):
        print("Flour determination: ", r)
    
    # Which - if any - sweetener?
    r = random.random()    
    if r < .1:
        ingredients[11] = 1
    elif r < .2:
        ingredients[12] = 1
    elif r < .8:
        ingredients[5] = 1
        
    if (verbose):
        print("Sweetener determination: ", r)
    
    # Which - if any - lipids?
    r = random.random()    
    if r < .25:
        ingredients[9] = 1
    elif r < .5:
        ingredients[10] = 1
    elif r < .75:
        ingredients[4] = 1
        
    if (verbose):
        print("Lipid determination: ", r)
    
    # Which - if any - other ingredients?
    
    # Salt? (All recipes start with salt, but distribution can push to 0; will not be below 0)
    ingredients[2] = 1
        
    # Egg?
    r = random.random()    
    if r < .2:
        ingredients[6] = 1
    if (verbose):
        print("Egg determination: ", r)
        
    # Dry milk?
    r = random.random()    
    if r < .1:
        ingredients[7] = 1
    if (verbose):
        print("Dry milk determination: ", r)
        
    if (verbose):
        print("Array without masses: \n", ingredients)
        
    ingredients[0] = random.random() * 12.5 + 2.5
    
    for i in range(1, 14):
        if (ingredients[i] > 0):
            ingredients[i] = numpy.random.normal(ing_averages[i], ing_stdev[i])
            if (ingredients[i]) < 0:
                ingredients[i] = 0

    return numpy.asarray(ingredients)

```

```python
for i in range(5):
    t = generateRecipe(False)
    print(t, "\n")
```
